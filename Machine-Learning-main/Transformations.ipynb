```python
import numpy as np
import statsmodels.api as sm
import pylab as py
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data_points = np.random.normal(0,1,100)
sm.qqplot(data_points, line = '45')
py.show()
```

```python
sns.displot(data_points, kind = 'kde')
py.show()
```

```python
gfg = np.random.lognormal(0,1,100)
print(gfg)
```

```python
sm.qqplot(np.log(gfg), line = '45')
py.show()
```

```python
sns.displot(np.log(gfg), kind = 'kde')
```

```python
'''Converting a non-normal distribution into normal distribution by using box-cox tranform'''
original_data = np.random.exponential(size = 10000)#create aoreto distribution
fitted_data, fitted_lambda = stats.boxcox(original_data)

fig, ax = plt.subplots(1,2)

sns.distplot(original_data, hist = False, kde = True, kde_kws = {'shade': True,
                                                                'linewidth': 2},
            label = 'Non-Normal', color = 'blue', ax = ax[0])

sns.distplot(fitted_data, hist = False, kde = True, kde_kws = {'shade': True,
                                                              'linewidth':2},
            label = 'normal', color = 'blue', ax = ax[1])

plt.legend(loc = 'upper right')

fig.set_figheight(3)
fig.set_figwidth(10)
print(f"Lambda value used for transformation: {fitted_lambda}")


```

```python
sm.qqplot(original_data, line = '45')
py.show()
```

```python
sm.qqplot(np.log(original_data), line = '45')
py.show()
```

```python
sm.qqplot(fitted_data, line = '45')
py.show()
```
